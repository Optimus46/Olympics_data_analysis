```python
data = pd.read_csv('CompanyProfit.csv')
data2 = pd.read_csv('CompanyProfit_new.csv')
data3 = pd.read_csv('CompanySuccess.csv')
```

```python
data.head(10)
```

```python
data.isnull().sum()
```

```python
import seaborn as sns
```

```python
sns.displot(data['Profit ($)'])
```

```python
#fill na using mean
col = data['Profit ($)']

col = col.fillna(col.mean(), inplace = True)
```

```python
data.isnull().sum()
```

```python
data2.isnull().sum()
```

```python

sns.displot(data2['Profit ($)'])
```

```python
#since it is rightly skewed it has outliers so use median to fill na
col = data2['Profit ($)']

col = col.fillna(col.median(), inplace = True)
```

```python
# Mean - For Normal Distribution
# Median - Right or Left Skewed Data   (Not Normal Distribution)
```
