```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Load data (update filepath if needed)
df = pd.read_csv('index.csv')
```

```python
# Preprocessing: Fix column names, parse dates
df.columns = [col.strip().lower() for col in df.columns]
df['date'] = pd.to_datetime(df['date'], errors='coerce')
df = df.sort_values('date')
```

```python
handling missing values
```

```python



# 1. Drop rows with any missing values
df_cleaned = df.dropna()

# 2. Fill missing values with a constant (e.g., 0 or 'Unknown')
df_filled_constant = df.fillna({'money': 0, 'coffee_name': 'Unknown'})

# Forward fill missing values
df_filled_ffill = df.ffill()

# Backward fill missing values
df_filled_bfill = df.bfill()


# Convert 'date' column to datetime if not already done
df['date'] = pd.to_datetime(df['date'])

# Set 'date' as index
df = df.set_index('date')

# Now apply time interpolation on 'money'
df['money'] = df['money'].interpolate(method='time')

# If needed, reset index later
df = df.reset_index()

# After handling missing values, verify no missing data remains
print(df.isnull().sum())
```

```python
df.columns
```

```python
df['card'] = df['card'].fillna('Unknown')

```

```python
# Calculate the mode of the 'card' column
mode_value = df['card'].mode()[0]

# Fill missing values with the mode
df['card'] = df['card'].fillna(mode_value)
```

```python
# Calculate Q1 (25th percentile) and Q3 (75th percentile)
Q1 = df['money'].quantile(0.25)
Q3 = df['money'].quantile(0.75)
IQR = Q3 - Q1

# Define outliers as points outside 1.5 * IQR from Q1 and Q3
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

# Filter the dataframe to keep only non-outliers
df_no_outliers = df[(df['money'] >= lower_bound) & (df['money'] <= upper_bound)]

print(f"Removed {len(df) - len(df_no_outliers)} outliers from 'money' column")
```

```python
# Daily sales
daily_sales = df.groupby('date')['money'].sum()
plt.figure(figsize=(12, 5))
daily_sales.plot()
plt.title('Daily Sales Over Time')
plt.ylabel('Sales Amount')
plt.show()
```

```python
# Weekly sales
weekly_sales = daily_sales.resample('W').sum()
plt.figure(figsize=(12, 5))
weekly_sales.plot()
plt.title('Weekly Sales Trend')
plt.ylabel('Sales Amount')
plt.show()
```

```python
# Sales by coffee type
plt.figure(figsize=(10, 4))
sns.barplot(x='coffee_name', y='money', data=df, errorbar=None, estimator=sum)
plt.title('Total Sales by Coffee Type')
plt.xticks(rotation=45)
plt.show()
```

```python
# Ensure  'date' column is datetime dtype
df['date'] = pd.to_datetime(df['date'])

# Aggregate daily sales
daily_sales = df.groupby('date')['money'].sum()

# Set frequency to daily  this tells pandas and statsmodels your data is daily
daily_sales = daily_sales.asfreq('D')

# Now fit ARIMA model
from statsmodels.tsa.arima.model import ARIMA
model = ARIMA(daily_sales, order=(2, 1, 2))
results = model.fit()

# Forecast
next_day = results.forecast(steps=1)
next_week = results.forecast(steps=7).sum()
next_month = results.forecast(steps=30).sum()

print("Next Day Sales Forecast:", next_day.values[0])
print("Next Week Sales Forecast:", next_week)
print("Next Month Sales Forecast:", next_month)
```

```python
Sales Forecasting Insights
The ARIMA model (order (2,1,2)) was applied on daily aggregated sales data, providing a quantitative prediction of future sales volumes.

The next day sales forecast (~354) gives a short-term estimate of expected revenue, useful for daily inventory and staffing planning.

Aggregated forecasts for the next week (~3202) and next month (~14311) help in medium- to longer-term demand planning and resource allocation.

Model and Practical Implications
Despite initial warnings about missing frequency information in the date index, the forecast remains interpretable and offers actionable insights.

Forecast estimates indicate expected demand fluctuations, which can guide procurement, marketing efforts, and cash flow management.

Regular updates and retraining of the model with new data will improve accuracy and responsiveness to changing sales patterns.

Limitations and Considerations
The predictions reflect historical patterns and assume continuity in factors affecting sales; unusual events or market changes may impact accuracy.

Ensuring proper time series frequency indexing and handling missing dates can improve forecast reliability.

Complementary approaches like including seasonality explicitly or using machine learning models may enhance predictions.
```
